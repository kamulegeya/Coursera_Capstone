```python
import pandas as pd
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot') # optional: for ggplot-like style
```

```python
df_export=pd.read_csv(r'D:\Docs\coffee_data\exports-crop-year.csv')
```

```python
df_export.columns
```

```python
## Here i put columns in a list and create a seperate list for the year variable
## just one line of code
cols=list(df_export.columns)[1:]
cols
```

```python
##use melt function to turn data into a normalised format
df = pd.melt(df_export, id_vars='exports_crop_year', value_vars=cols)
### we put column namws 
col_n=["Country","Year","Qty"]
df.columns=col_n
df.head()
```

```python
##east africa
east=['Uganda','Kenya','Tanzania','Rwanda','Ethiopia']
east_df=[]
for c in east:
    #put name
    #name='df_'+c
    #print(name)  
    east_df.append(df[(df['Country']==c)])  
```

```python
#east_df[0]
```

```python

def get_color(c):
        switcher={
                'Uganda':'red',
                'Tanzania':'green',
                'Rwanda':'blue',
                'Kenya':'orange',
                'Ethiopia':'Purple'
                
             }
        return switcher.get(c)
```

```python
plt.figure(figsize=(12,10))
for i in range(0,len(east_df)):
    plt.plot(east_df[i]['Year'],east_df[i]['Qty'],c= get_color(east[i]),label=east[i])
    leg = plt.legend(loc='best', ncol=3 , shadow=True, fancybox=True)
    plt.title('East Africa  Coffee Exports  Trends Between 1990-2018(Source ICO Dataset)')
    plt.xlabel("Year")
    plt.ylabel("Thousand 60kg Bags")
    plt.xticks(rotation=45)
    leg.get_frame().set_alpha(0.5)
    plt.savefig(r'D:\Workings\east_coffeeexports.png',dpi=300,transparent=False)
```
